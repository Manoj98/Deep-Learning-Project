```python
import pandas as pd
import json
```

```python
with open('../data/synthetic_answers_squad_copy.txt') as f: 
    data = f.read()
js = json.loads(data)
```

```python
js
```

```python
len(js)
```

```python
temp = {}
for key in js.keys():
  temp[key] = js[key].split('\n')
```

```python
temp
```

```python
final = {}
for key in temp.keys():
  try:
    final[int(key)] = [x.split('. ')[1] if x != 'Answers: ' else x for x in temp[key] if x != 'Answers:']
  except:
    print(key)
    final[int(key)] = temp[key]
```

```python
for key in final.keys():
  final[key] = [x.replace('Answer: ', '') for x in final[key]]
```

```python
final
```

```python
len(final)
```

```python
## Saving processed synthetic data
with open('synthetic_processed.json', 'w', encoding='utf-8') as f:
    json.dump(final, f, ensure_ascii=False, indent=4)
```

```python
## Creating dataframe for original data
with open('data_original.json','r') as f:
    data = json.loads(f.read())
    
## Normalizing data
df = pd.json_normalize(data, record_path =['records'], meta=['context'])
```
